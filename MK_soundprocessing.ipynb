# [1] Import modules

```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
# parameter setting: amplitude, sampling rate, duration, frequency, phase

amp = 1           # range [0.0, 2.0]
sr = 10000        # sampling rate, Hz
dur = 0.5         # in seconds
freq = 150.0      # sine frequency, Hz
phase = 0  # phase, radian (e.g., 2pi)
```

## [2] Generate sample ticks, sample to time, time to angle, shift angle

```python
# generate sample ticks
nSamp = sr*dur
samples = np.arange(1, nSamp+1)
len(samples)
```

sr = Nsample/1 sec
t = Nsample / sr


```python
# Convert sample to time (sec)
t = samples/sr; t
```

### w = 2pi*freq

Screen Shot 2022-04-27 at 6.22.02 PM.png

```python
# Convert time to angle (radian): e.g., sin(angle), angle in theta
theta0 = t * 2*np.pi * freq
```

```python
# shift angle by phase
theta = theta0 + phase
```

```python
# Generate signal by sin-phasor
s = amp*np.sin(theta)
```

## [3] Plot the signal

```python
# Plot the signal
plt.plot(t,s,':')
```

```python
# plot a certain portion of signals
samplePart = np.arange(0,100)
plt.plot(t[samplePart],s[samplePart],':')
plt.xlabel('time (s)')
plt.ylabel('amp')
```

# [4] Trial: Creting a complex wave

```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
# parameter setting: amplitude, sampling rate, duration, frequency, phase
# Sound A
amp = 4           # range [0.0, 1.0]
sr = 10000        # sampling rate, Hz
dur = 1         # in seconds
freq = 100.0      # sine frequency, Hz
phase = 0  # phase, radian (e.g., 2pi)

# Sound B
amp1 = 2          # range [0.0, 0.5]
sr = 10000        # sampling rate, Hz
dur = 1         # in seconds
freq1 = 200.0      # sine frequency, Hz
phase = 0  # phase, radian (e.g., 2pi)

# Sound C
amp2 = 0.5          # range [0.0, 0.5]
sr = 10000        # sampling rate, Hz
dur = 1         # in seconds
freq2 = 150.0      # sine frequency, Hz
phase = 0  # phase, radian (e.g., 2pi)
```

```python
nSamp = sr*dur
samples = np.arange(1, nSamp+1)
len(samples)
```

```python
t = samples/sr
theta0 = t * 2*np.pi * freq; theta1 = t*2*np.pi * freq1; theta2 = t*2*np.pi * freq2
theta = theta0 + phase; theta3 = theta1 + phase; theta4 = theta2 + phase
s = amp*np.sin(theta); s1 = amp1*np.sin(theta3); s2 = amp2*np.sin(theta4)
```

```python
# Sound A: plot a certain portion of signals
samplePart = np.arange(0,100)
plt.plot(t[samplePart],s[samplePart],':')
plt.plot(t[samplePart],s1[samplePart],':')
#plt.plot(t[samplePart],s2[samplePart],':')
plt.plot(t[samplePart],(s[samplePart] + s1[samplePart] + s2[samplePart]),'-')
plt.xlabel('time (s)')
plt.ylabel('amp')
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr*dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp*np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; dur=1; freq=220; phrase=0
s, t, theta = phasor_sine(amp, sr, dur, freq, phase)
ipd.Audio(s, rate=sr)
```
