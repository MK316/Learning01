```python
!pip install -q bitsandbytes datasets loralib sentencepiece git+https://github.com/huggingface/transformers.git git+https://github.com/huggingface/peft.git gradio==3.20.0 tenacity
```

```python
from peft import PeftModel
from transformers import LlamaTokenizer, LlamaForCausalLM

base = 'decapoda-research/llama-13b-hf'
finetuned = 'beomi/KoAlpaca-13B-LoRA'
# finetuned = 'chansung/koalpaca-lora-13b'

tokenizer = LlamaTokenizer.from_pretrained(base)
tokenizer.pad_token_id = 0
tokenizer.padding_side = "left"

model = LlamaForCausalLM.from_pretrained(
    base,
    load_in_8bit=True,
    device_map="auto",
)

model = PeftModel.from_pretrained(model, finetuned, device_map={'': 0})
```

- Instruction: str, 모델이 수행해야하는 태스크를 설명한다.
- Input: (선택)str, 태스크를 수행하기위한 입력. 예) 이 글을 요약해주세요 라는 인스트럭션이 있다면 입력으로 "글"을 입력한다.
- Response: str, 답변

```python
input = ""
prompt = "소수(Prime number)를 순서대로 출력하는 파이썬 코드를 작성해주세요."

history = f"""Below is a history of instructions that describe tasks, paired with an input that provides further context. Write a response that appropriately completes the request by remembering the conversation history.
아래는 작업을 설명하는 지시 사항의 히스토리와 추가 컨텍스트를 제공하는 입력이 짝을 이루고 있습니다. 대화 기록을 기억하여 요청을 적절히 완료하는 응답을 작성하세요.

### Input: {input}

### Instruction: {prompt}

### Response:"""

print(history)
```

```python
max_new_tokens = 128
temperature = 0.5

input_ids = tokenizer.encode(history, return_tensors="pt").to('cuda:0')

gen_tokens = model.generate(
    input_ids=input_ids, 
    max_new_tokens=max_new_tokens, 
    num_return_sequences=1, 
    temperature=temperature,
    no_repeat_ngram_size=6,
    do_sample=True,
)

gen_text = tokenizer.decode(gen_tokens[0], skip_special_tokens=True)

print(gen_text.replace(history, ''))
```

```python
def is_prime(n):
    if n == 2:
        return True
    if n ==1:
        return False
    if n%2 == 0:
        return False # 2가 나열 숫자를 찾으면 반올림
    for i in range(3, int(n**0.5)+1, 2):
        if n%i == 0:
            return False
    return True

for i in range(30):
    if is_prime(i):
        print(i)
```

```python
prompt = "위 코드를 파이썬으로 변환해주세요."

history = gen_text + f"""

### Instruction: {prompt}

### Response:"""

print(history)
```

```python
for i in range(2, 10):
    if i % 2 == 0:
        continue
    else:
        print(i)
```

```python
input_ids = tokenizer.encode(history, return_tensors="pt").to('cuda:0')

gen_tokens = model.generate(
    input_ids=input_ids, 
    max_new_tokens=max_new_tokens, 
    num_return_sequences=1, 
    temperature=temperature,
    no_repeat_ngram_size=6,
    do_sample=True,
)

gen_text = tokenizer.decode(gen_tokens[0], skip_special_tokens=True)

print(gen_text.replace(history, ''))
```

## 2. GPT4All

Colab 유료 버전 - 고용량 RAM 사용시 실행 가능

- https://github.com/nomic-ai/gpt4all

```python
!pip install -q bitsandbytes datasets loralib sentencepiece git+https://github.com/huggingface/transformers.git git+https://github.com/huggingface/peft.git gradio==3.20.0 tenacity
```

```python
import torch
from peft import PeftModel, PeftConfig
from transformers import AutoModelForCausalLM, AutoTokenizer, GenerationConfig
import textwrap

peft_model_id = "nomic-ai/gpt4all-lora"
config = PeftConfig.from_pretrained(peft_model_id)

model = AutoModelForCausalLM.from_pretrained(
    config.base_model_name_or_path, 
    return_dict=True, 
    load_in_8bit=True, 
    device_map='auto')

tokenizer = AutoTokenizer.from_pretrained(config.base_model_name_or_path)

model = PeftModel.from_pretrained(model, peft_model_id)
```

```python
def gpt4all_generate(text):
    inputs = tokenizer(
        text,
        return_tensors="pt",
    )
    input_ids = inputs["input_ids"].cuda()

    generation_config = GenerationConfig(
        temperature=0.6,
        top_p=0.95,
        repetition_penalty=1.2,
    )

    print("Generating...")
    generation_output = model.generate(
        input_ids=input_ids,
        generation_config=generation_config,
        # return_dict_in_generate=True,
        output_scores=True,
        max_new_tokens=256,
    )

    wrapped_text = textwrap.fill(tokenizer.decode(generation_output[0]), width=100)
    
    return wrapped_text
```

```python
prompt = "Tell me about yourself."

history = f"""Below is a history of instructions that describe tasks, paired with an input that provides further context. Write a response that appropriately completes the request by remembering the conversation history.

### Instruction: {prompt}

### Response:"""

gen_text = gpt4all_generate(history)

print(gen_text.replace(history, ''))
```

```python
prompt = "한글로 번역해주세요."

history = gen_text + f"""

### Instruction: {prompt}

### Response:"""

gen_text = gpt4all_generate(history)

print(gen_text.replace(history, ''))
```
